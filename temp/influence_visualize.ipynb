```python
# Load trained_models/leave_out_base_skew_major.npy
import numpy as np

lo_mj_base = np.load("../trained_models/leave_out_base_skew_major.npy").flatten()
lo_mn_base = np.load("../trained_models/leave_out_base_skew_minor.npy").flatten()
lo_mj_syn = np.load("../trained_models/leave_out_syn_skew_major.npy").flatten()
lo_mn_syn = np.load("../trained_models/leave_out_syn_skew_minor.npy").flatten()

mj_base = np.load("../trained_models/base_skew_major.npy").flatten()
mn_base = np.load("../trained_models/base_skew_minor.npy").flatten()
mj_syn = np.load("../trained_models/syn_skew_major.npy").flatten()
mn_syn = np.load("../trained_models/syn_skew_minor.npy").flatten()
```

```python
print(lo_mj_base.shape, lo_mn_base.shape, lo_mj_syn.shape, lo_mn_syn.shape)
print(mj_base.shape, mn_base.shape, mj_syn.shape, mn_syn.shape)
```

```python
# Draw distribution of lo_mj_base, lo_mn_base
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.hist(lo_mj_base, bins=30, alpha=0.7, label='Leave-One-Out Majority Base', color='blue')
plt.hist(lo_mn_base, bins=30, alpha=0.7, label='Leave-One-Out Minority Base', color='red')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution of Leave-One-Out Majority Base vs Leave-One-Out Minority Base')
# Write the caption show mean and variance of lo_mj_base and lo_mn_base
plt.figtext(0.15, 0.8, f'LO_MJ_BASE Mean: {lo_mj_base.mean():.2f}, Var: {lo_mj_base.var():.2f}', fontsize=12, color='blue')
plt.figtext(0.15, 0.75, f'LO_MN_BASE Mean: {lo_mn_base.mean():.2f}, Var: {lo_mn_base.var():.2f}', fontsize=12, color='red')
plt.legend()
plt.grid(True)
plt.show()
```

```python
dr_mn_if = (lo_mn_base[:, None] - mn_base[None, :]).flatten()
dr_mj_if = (lo_mj_base[:, None] - mj_base[None, :]).flatten()
idr_mn_if = (lo_mn_syn[:, None] - mn_syn[None, :]).flatten()
idr_mj_if = (lo_mj_syn[:, None] - mj_syn[None, :]).flatten()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.hist(dr_mj_if, bins=30, alpha=0.7, label='Majority Direct Influence', color='blue')
plt.hist(dr_mn_if, bins=30, alpha=0.7, label='Minority Direct Influence', color='red')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution of Majority Direct Influence vs Minority Direct Influence')
# Write the caption show mean and variance of lo_mj_base and lo_mn_base
plt.figtext(0.15, 0.8, f'Majority Direct Influence Mean: {dr_mj_if.mean():.2f}, Var: {dr_mj_if.var():.2f}', fontsize=12, color='blue')
plt.figtext(0.15, 0.75, f'Minority Direct Influence Mean: {dr_mn_if.mean():.2f}, Var: {dr_mn_if.var():.2f}', fontsize=12, color='red')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.hist(idr_mj_if, bins=30, alpha=0.7, label='Majority Indirect Influence', color='blue')
plt.hist(idr_mn_if, bins=30, alpha=0.7, label='Minority Indirect Influence', color='red')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution of Majority Indirect Influence vs Minority Indirect Influence')
# Write the caption show mean and variance of lo_mj_base and lo_mn_base
plt.figtext(0.15, 0.8, f'Majority Indirect Influence Mean: {idr_mj_if.mean():.2f}, Var: {idr_mj_if.var():.2f}', fontsize=12, color='blue')
plt.figtext(0.15, 0.75, f'Minority Indirect Influence Mean: {idr_mn_if.mean():.2f}, Var: {idr_mn_if.var():.2f}', fontsize=12, color='red')
plt.legend()
plt.grid(True)
plt.show()
```
